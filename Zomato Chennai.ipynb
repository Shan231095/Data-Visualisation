## Q2) What are the most popular top dishes in Chennai restaurants?

```python

top_dishes_df = zomato_df['Top Dishes'].value_counts().reset_index()
top_dishes_df = top_dishes_df[top_dishes_df['Top Dishes'].astype(str) != '0']
top_dishes_df.rename(columns={'index': 'Top Dishes', 'Top Dishes': 'Top Dishes of Chennai'}, inplace=True) 
print(top_dishes_df.head(10))
```

## Q3) Which locations in Chennai have the most restaurants?

```python
location_counts = raw_df.groupby('Location')['Name of Restaurant'].count().reset_index()
location_counts = location_counts.rename(columns={'Name of Restaurant': 'Restaurant Count'})
location_counts = location_counts.sort_values(by='Restaurant Count', ascending=False)

print(location_counts.head(10))
```

```python
top_locations = location_counts.head(10)
plt.figure(figsize=(12,6))
plt.barh(top_locations['Location'], top_locations['Restaurant Count'], color='skyblue')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.title('Top 10 Locations in Chennai with Most Restaurants')
plt.gca().invert_yaxis()  # To display highest at the top
plt.show()
```

## Q7) Which restaurants have high delivery ratings but low dining ratings?

```python
# Define thresholds 
high_delivery_threshold = 4.0  # Consider ratings above 4.0 as high
low_dining_threshold = 3.0     # Consider ratings below 3.0 as low

# Convert ratings to numeric (handle non-numeric values like 'NEW' or '-')
raw_df['Dining Rating'] = pd.to_numeric(raw_df['Dining Rating'], errors='coerce')
raw_df['Delivery Rating'] = pd.to_numeric(raw_df['Delivery Rating'], errors='coerce')

# Filter restaurants meeting the condition
high_delivery_low_dining = raw_df[(raw_df['Delivery Rating'] >= high_delivery_threshold) & 
                                  (raw_df['Dining Rating'] < low_dining_threshold)]

# Select relevant columns
high_delivery_low_dining = high_delivery_low_dining[['Name of Restaurant', 'Location', 'Dining Rating', 'Delivery Rating']]

# Display the first few results
print(high_delivery_low_dining.head(10))
```

## Q8) Which locations have the cheapest dining options?

```python
# Group by 'Location' and calculate the average price
cheapest_locations = raw_df.groupby('Location')['Price for 2'].mean().reset_index()

# Rename the column for clarity
cheapest_locations.rename(columns={'Price for 2': 'Average Price for 2'}, inplace=True)

# Sort by lowest average price
cheapest_locations = cheapest_locations.sort_values(by='Average Price for 2', ascending=True)

# Display the top 10 cheapest locations
print(cheapest_locations.head(10))
```

## Q9) Which restaurant features (e.g., live music, rooftop dining, parking) are most common?

```python
# Convert comma-separated feature lists into separate rows
features_series = raw_df['Features'].str.split(',').explode().str.strip()

# Count occurrences of each feature
feature_counts = features_series.value_counts()

# Remove unwanted characters from feature names
feature_counts_cleaned = feature_counts.rename(lambda x: x.replace("[", "").replace("]", "").replace("'", "").strip())

# Display cleaned feature counts
print(feature_counts_cleaned.head(10))
```

## Q10) Find the list of Affordable Restaurants?
##### The criteria for Affordable Restaurants would be:
##### 1) Low Price 2) High Rated
##### First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
##### Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
##### restaurants that are 1/4th of 6000.

```python
max_price = zomato_df['Price for 2'].max() 
one_fourth_price = max_price/4 
one_fourth_price 
```

```python
zomato_df.columns
```

```python
# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restaurant
 
aff_rest_df = zomato_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Features']] 
aff_rest_df = aff_rest_df[aff_rest_df['Price for 2'] <= 750] 
aff_rest_df.sort_values(by='Price for 2', inplace=True) 
aff_rest_df.reset_index(drop=True, inplace=True)
aff_rest_df 
```

```python
# Finding the highest rated list of restaurants 
 
highrate_rest_df = zomato_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Features','Dining Rating']] 
highrate_rest_df = highrate_rest_df[highrate_rest_df['Dining Rating'] >= 4.5] 
highrate_rest_df.sort_values(by='Price for 2', inplace=True) 
highrate_rest_df.head(10)
```

```python
zomato_df.columns
```

#### Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !!

```python
highrate_aff_df = pd.merge(aff_rest_df, highrate_rest_df, how='inner', on=['Name of Restaurant', 'Location']) 
highrate_aff_df = highrate_aff_df[['Name of Restaurant', 'Price for 2_x', 'Cuisine_x', 'Location', 'Features_x']] 
highrate_aff_df.rename(columns={'Name of Restaurant':'Name of Restaurant', 'Price for 2_x':'Price for 2', 'Cuisine_x':'Cuisine',  
                                'Location':'Location', 'Features_x':'Features'},inplace=True)
```

```python
# Affordable Restaurants with low price and high rating  
 
highrate_aff_df 
```

##  Q11) Find the list of most Reliable Restaurants?
#### The criteria for most Reliable Restaurants would be:
#### 1) Low Price 2) High Rated 3) Large No. of Votes
#### First step will be to find the restaurants with Votes greater than Mean of Votes

```python
mean_votes = zomato_df['Dining Rating Count'].mean() 
mean_votes 
```

```python
# Finding list of restaurants that have Votes greater than and equal to  Mean of Vote 
 
mean_rest_df = zomato_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Features', 'Dining Rating Count']] 
mean_rest_df = mean_rest_df[mean_rest_df['Dining Rating Count'] > 167] 
mean_rest_df.sort_values(by='Dining Rating Count', inplace=True) 
mean_rest_df 
```

##### These are the most reliable, highest rated and affordable restaurants:
##### We obtain this dataframe by simply taking the intersection of highrate_aff_df & mean_rest_df
##### This dataframe obtained below shows the restaurants whose:
##### Cost is below 1250
##### Rating is above 4.5
##### Votes are above 177

```python
reliable_rest_df = pd.merge(mean_rest_df, highrate_aff_df, how='inner', on=['Name of Restaurant', 'Location']) 
reliable_rest_df = reliable_rest_df[['Name of Restaurant', 'Price for 2_x', 'Cuisine_x', 'Location', 'Features_x']] 
reliable_rest_df.rename(columns={'Name of Restaurant':'Name of Restaurant', 'Price for 2_x':'Price for 2', 'Cuisine_x':'Cuisine',  
                                'Location':'Location', 'Featues_x':'Features'},inplace=True)
```
